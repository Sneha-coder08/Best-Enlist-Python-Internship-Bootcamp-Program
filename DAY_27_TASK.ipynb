## **Take any webpage, Scrape the data and try to insert only the class data and save it in excel**

```python
import csv
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
from bs4 import BeautifulSoup
import requests
url='https://islamqa.info/en/answers/1/interruption-of-wudhu'
html=requests.get(url)
soup=BeautifulSoup(html.text, "html.parser")
print(soup.prettify())
```

```python
txt=soup.findAll(attrs={"class":"content"})
print(txt)
```

```python
Question=txt[0].text.replace('\n'," ")
Summary=soup.find(attrs={"class":"title is-4 is-size-5-touch"}).text.replace('\n'," ")
QuestionNo=int(soup.find(attrs={"class":"subtitle has-text-weight-bold has-title-case cursor-pointer tooltip"}).text.replace('\n'," "))
Source=soup.find(attrs={"class":"subtitle is-6 has-text-weight-bold is-capitalized"}).text.replace('\n'," ")
```

```python
data=[[Question,Summary,QuestionNo,Source]]
```

```python
data
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["Question","Summary","QuestionNo","Source"])
```

```python
for i in range(1,10):
    URL="https://islamqa.info/en/answers/1/interruption-of-wudhu"
    page=requests.get(URL)
    if(page.status_code==200):
        print("Data Fetched Successfully",i)
        soup=BeautifulSoup(page.content, "html.parser")
        txt=soup.findAll(attrs={"class":"content"})
        Q=txt[0].text.replace('\n'," ")
        S=soup.find(attrs={"class":"title is-4 is-size-5-touch"}).text.replace('\n'," ")
        QN=int(soup.find(attrs={"class":"subtitle has-text-weight-bold has-title-case cursor-pointer tooltip"}).text.replace('\n'," "))
        So=soup.find(attrs={"class":"subtitle is-6 has-text-weight-bold is-capitalized"}).text.replace('\n'," ")
        data.insert(QN,[QN,S,Q,So])
    else:
        print("URL Not Found",i)
df=pd.DataFrame(data,columns=["Question","Summary","QuestionNo","Source"])
df.to_csv("Data.csv")
```
